```python
df_dict = {'beer_name': [], 'beer_id': [],
           'date': [], 'user_name': [], 'user_id': [],
           'overall': [], 'text': []}
with open("Data/BeerAdvocate/ratings.txt") as f:
  for i,line in enumerate(f):
    if line != '\n':
      key, value = line.rstrip().split(":", maxsplit=1)
      if key in df_dict.keys():
        df_dict[key].append(value.strip())
    if i>=18*3_000_000-1:   
      break
ratings_ba1 = pd.DataFrame.from_dict(df_dict)
indx_to_drop = ratings_ba1[(ratings_ba1['text']=='nan') | (ratings_ba1['overall']=='nan')].index
ratings_ba1 = ratings_ba1.drop(indx_to_drop)
ratings_ba1 = ratings_ba1.reset_index()
ratings_ba1
```
